## Definition
#### Artificial Neural Network(ANN) are computing systems isnpired by the bilogical neural networks that constitute animal brains. The nueral network is not an algorithm but rather a framework for many different machine learning algorithms to work together and process complex data inputs. Such system learn to perform task by considering examples, generally without being programmed with any task-specific rules.
<img src='nueral.JPG'>

#### For example, in image recognition, they might learn to identify images that contain cats by analyzing example images that have been manually labeled as “cat” or “no cat” and using the results to identify cats in other images. They do this without any prior knowledge about cats, e.g., that they have fur, tails, whiskers and cat-like faces. Instead, they automatically generate identifying characteristics from the learning material that they process.

#### An ANN is based on a collection of connected units or nodes called artificial neurons, which loosely model the neurons in a biological brain. Each connection, like the synapses in a biological brain, can transmit a signal from one artificial neuron to another. An artificial neuron that receives a signal can process it and then signal additional artificial neurons connected to it.

### What is Back Propagation?
- Backpropagation is the central mechanism by which neural networks learn. It is the messenger telling the network whether or not the net made a mistake when it made a prediction. ... Forward propagation is when a data instance sends its signal through a network's parameters toward the prediction at the end.
- Back-propagation is just a way of propagating the total loss back into the neural network to know how much of the loss every node is responsible for, and subsequently updating the weights in such a way that minimizes the loss by giving the nodes with higher error rates lower weights and vice versa

### How does back propagation work?
The algorithm is used to effectively train a neural network through a method called `Chain Rule`. In simple terms, after each forward pass through a network, backpropagation performs a backward pass while adjusting the model's parameters (weights and biases).

### Task: To preadict Person will stay with Bank or Not?

```python
X = data.drop(labels=['CustomerId','Surname','Exited','RowNumber'],axis=1)
y = data['Exited']
```

```python
X.head()
```

```python
y.head()
```

ANN WORKS ON NUMERICAL DATA NOT ON STRING DATA HERE AS YOU CAN SEE GEOGRAPHY AND GENDER ARE STRING SO WE CONVERT IT INTO NUMERICAL VALUE THAT MEANS WE HAVE TO MAP THESE LABLE IN TO NUMERICAL VALUES BUT HERE I USE LABEL ENCODER AND ONE HOT ENCODER TO CONVERT THESE STRING LABELS INTO NUMERICAL VALUES.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label1 = LabelEncoder()
X['Geography'] = label1.fit_transform(X['Geography'])
```

```python
X.head(10)
```

```python
label2 = LabelEncoder()
X['Gender'] = label2.fit_transform(X['Gender'])
```

```python
X.head(10)
```

```python
X = pd.get_dummies(X, drop_first=True, columns=['Geography'])
```

```python
X.head()
```

### Feature Standardization
Scales are varying too much so in machine learning variance differences between the columns is not recomanded so what we need to bring out all features at the same scalefor this we use Standard scaler.  

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=0,stratify=y)
```

```python
Scaler = StandardScaler()
X_train = Scaler.fit_transform(X_train)
X_test = Scaler.fit_transform(X_test)
```

```python
X_train
```

### Build ANN

```python
X.shape
```

```python
model = Sequential()
# Input Layer
model.add(Dense(X.shape[1],activation='relu',input_dim = X.shape[1]))
# HiddenLayer
model.add(Dense(128,activation='relu'))
# output layer
model.add(Dense(1,activation='sigmoid'))
```

### compile our model

```python
# optimizer = 'adam',loss function = binary_cross Entropy,
# ON which basis model is evaluated
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# batch size = 1 means for each input weight will change
# batch size = total data then stochastic gradient will become batch gradient
model.fit(X_train,y_train.to_numpy(),batch_size=10,epochs = 10,verbose=1)
```

### Termilology:
This 86.3% accuracy based on training data sets not based on test data

### Predict classes

```python
y_pred = model.predict_classes(X_test)
```

```python
# predicted y
y_pred
```

```python
# actual y
y_test
```

```python
model.evaluate(X_test,y_test.to_numpy())
```
